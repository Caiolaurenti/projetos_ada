# Aulas 02 - Matplotlib

O matplotlib é uma biblioteca que permite graficar diversos tipos de representações diretamente a partir de arrays. Possui essencialmente todos os tipos de gráficos que estudamos na aula passada, mas com uma sintaxe diferente a do pandas, o que afetará não somente como plotamos as visualizações, mas, também, elementos "estéticos" associados a elas (como legendas, eixos e afins). 

A biblioteca do matplotlib define alguns **elementos básicos** a partir dos quais se estruturam todas as suas funções.

### Elementos básicos de uma figura com matplotlib

[Guia oficial do matplotlib](https://matplotlib.org/stable/tutorials/introductory/quick_start.html)

Na figura abaixo, temos definidos os **elementos básicos** mencionados acima. Eles servirão como parâmetros nas mais variadas funções da biblioteca.

<img src="https://matplotlib.org/stable/_images/anatomy.png" alt="drawing" width="600"/>

É essencial que entendamos três outros elementos fundamentais para a melhor utilização da biblioteca matplotlib:

**-Figure:** é a imagem como um todo; a "figura" prorpiamente dita.

**-Axes:** contém todos os elementos do plot em si, incluindo os eixos, mas não apenas eles.

**-Axis:** são os eixos do gráfico (eixo-x; eixo-y; eixo-z).

Voltando ao nosso conjunto de dados...

```python
df.head()
```

Vamos fazer um gráfico de barras, como acima, mas por ano.

```python
df['date'].min()
```

```python
df['date'].max()
```

```python
# Vamos dividir nossa figura em 4 gráficos (um por ano)
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (14,10))
ax = np.reshape(ax, (1,4))

year = ['2011', '2012', '2013', '2014']

subplot_counter = 1
for i in range(0,4):
    plt.subplot(2,2,subplot_counter)
    
    # Para cada ano
    df_year = df[(df['date'] >= year[i]+'-01-01') & (df['date'] <= year[i]+'-12-31')]
    df_grouped = df_year.groupby("State")['Profit'].sum().sort_values(ascending = False)
    
    x = list(df_grouped.index)
    y = df_grouped.values
    
    bars = plt.bar(x,y)
    subplot_counter += 1
    
    plt.title(year[i])
    
    ax[0,i].bar_label(bars, fmt = '%.0f')

    ax[0,i].spines['right'].set_visible(False)
    ax[0,i].spines['top'].set_visible(False)
    ax[0,i].spines['left'].set_visible(False)

    ax[0,i].get_yaxis().set_visible(False)

    ax[0,i].set_xticklabels(labels = x, rotation = 45)

plt.suptitle("Lucro anual por estado")
plt.tight_layout()
```

### Gerando algumas visualizações descritivas com matplotlib

Vamos utilizar um dataset [dos preços dos combustíveis no Brasil](https://www.kaggle.com/datasets/matheusfreitag/gas-prices-in-brazil).

Reproduza as visualizações abaixo. Para cada uma, pense em pontos que você poderia modificar na visualização, pensando em aprimorá-la.

#### Preços médios do etanol e da gasolina, por região do Brasil

```python
df = pd.read_csv('2004-2021.tsv', sep = '\t')
```

```python
df['DATA FINAL'] = [datetime.date() for datetime in pd.to_datetime(df['DATA FINAL'])]
```

```python
df_etanol = df[df['PRODUTO'] == 'ETANOL HIDRATADO']
df_gasolina = df[df['PRODUTO'] == 'GASOLINA COMUM']
```

```python
media_etanol = df_etanol.groupby(['REGIÃO', 'DATA FINAL'])['PREÇO MÉDIO REVENDA'].mean()
media_gasolina = df_gasolina.groupby(['REGIÃO', 'DATA FINAL'])['PREÇO MÉDIO REVENDA'].mean()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7.5))
x = np.sort(df['DATA FINAL'].unique())
regioes = df['REGIÃO'].unique()
for regiao in regioes:
    ax[0].plot(media_etanol[regiao].index, media_etanol[regiao], label = regiao)
    ax[0].legend()
    ax[0].set_xlim(x[0], x[-1])
    ax[0].set_ylim(0, 7)
    ax[0].xaxis.set_ticks(x[::50])
    ax[0].yaxis.set_ticks(range(8))
    ax[0].title.set_text('Média preço Etanol')
    ax[0].set_xticklabels(labels = x[::50], rotation = 45)
    ax[1].plot(media_gasolina[regiao].index, media_gasolina[regiao], label = regiao)
    ax[1].legend()
    ax[1].set_xlim(x[0], x[-1])
    ax[1].set_ylim(0, 7)
    ax[1].xaxis.set_ticks(x[::50])
    ax[1].yaxis.set_ticks(range(8))
    ax[1].title.set_text('Média preço Gasolina')
    ax[1].set_xticklabels(labels = x[::50], rotation = 45)
    ax[1].set_yticklabels(labels = range(8))
plt.show()
```

#### Preço médio em 2020 dos preços do Etanol e da Gasolina por região

```python
mask = [date.year == 2020 for date in df['DATA FINAL']]
df_2020 = df[mask]
```

```python
df_etanol_2020 = df_2020[df_2020['PRODUTO'] == 'ETANOL HIDRATADO']
df_gasolina_2020 = df_2020[df_2020['PRODUTO'] == 'GASOLINA COMUM']
```

```python
media_etanol_2020 = df_etanol_2020.groupby(['REGIÃO', 'DATA FINAL'])['PREÇO MÉDIO REVENDA'].mean()
media_gasolina_2020 = df_gasolina_2020.groupby(['REGIÃO', 'DATA FINAL'])['PREÇO MÉDIO REVENDA'].mean()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7.5))
time_range = np.arange(dt.date(year = 2020, month = 1, day = 1), dt.date(year = 2020, month = 12, day = 31),\
                       dt.timedelta(days = 1))
time_range = time_range.astype(dt.date)
time_range = [date.date() for date in time_range]
regioes = df_2020['REGIÃO'].unique()
for regiao in regioes:
    ax[0].plot(media_etanol_2020[regiao].index, media_etanol_2020[regiao], label = regiao)
    ax[0].legend()
    ax[0].set_xlim(time_range[0], time_range[-1])
    ax[0].xaxis.set_ticks(time_range[::40])
    ax[0].set_ylim(0, 7)
    ax[0].set_xticklabels(labels = time_range[::40], rotation = 45)
    ax[0].title.set_text('Média Etanol 2020')
    ax[1].plot(media_gasolina_2020[regiao].index, media_gasolina_2020[regiao], label = regiao)
    ax[1].legend()
    ax[1].set_xlim(time_range[0], time_range[-1])
    ax[1].xaxis.set_ticks(time_range[::40])
    ax[1].title.set_text('Média Gasolina 2020')
    ax[1].set_ylim(0, 7)
    ax[1].set_xticklabels(labels = time_range[::40], rotation = 45)
plt.show()
```

#### Relação entre preço da gasolina e do etanol (nível Brasil)

```python
df_filtrado = df[df['PRODUTO'].isin(['ETANOL HIDRATADO', 'GASOLINA COMUM'])]
grupos = df_filtrado.groupby(['DATA FINAL', 'PRODUTO'])['PREÇO MÉDIO REVENDA'].mean()
dates = sorted(list(set([tupla[0] for tupla in grupos.index if len(grupos[tupla[0]]) == 2])))
et = np.array([grupos[data]['ETANOL HIDRATADO'] for data in dates])
gas = np.array([grupos[data]['GASOLINA COMUM'] for data in dates])
rel = et/gas
padrao = np.repeat(0.7, len(rel))
```

```python
fig, ax = plt.subplots(figsize = (7, 4))
ax.set_xlim(dates[0], dates[-1])
ax.set_ylim(0.5, 1)
ax.xaxis.set_ticks(dates[::50])
ax.set_xticklabels(labels = dates[::50], rotation = 45)
ax.plot(dates, rel, label = 'ETANOL/GASOLINA')
ax.title.set_text('Evolução ETANOL/GASOLINA')
ax.plot(dates, padrao, linestyle = '--', label = 'Rendimento equivalente', color = 'r')
ax.legend()
plt.show()
```

A linha de rendimento acima serve para comparar a evolução de preço médio com o conhecimento popular de que "o álcool compensa se custa menos que 70% do preço da gasolina"
