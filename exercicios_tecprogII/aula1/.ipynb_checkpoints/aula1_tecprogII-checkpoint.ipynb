**Exercício 1.** Reproduza o gráfico de barras abaixo, em que cada barra representa um dos 10 países com mais casos confirmados de COVID no dataset, e a "quebra" em cada cor indica a predominância de casos confirmados em cada mês.

```python
df = pd.read_csv('Covid_19_Countrywise_timeseries.csv')
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
```

```python
df['Month'] = [dt.date(month = date.month, year = date.year, day = 1) for date in df['ObservationDate']]
```

```python
grouped = df.groupby(['Month', 'country'])['Confirmed'].sum()
```

```python
grouped = grouped.sort_values(ascending = False)
grouped = grouped[:10]
months = sorted(list(set([tupla[0] for tupla in grouped.index])))
```

```python
fig, ax = plt.subplots()
for month in months[::-1]:
    ax.bar(grouped[month].index, grouped[month], label = month)
    ax.legend()
    ax.set_xticklabels(labels = grouped[months[-1]].index, rotation = 45)

```

**Exercício 4.** Os 10 países que mais aumentaram, percentualmente, a emissão de CO2 entre os anos de 2018 e 2019, na forma de um gráfico de barras horizontal.

```python
import numpy as np
```

```python
df_2018 = df[df['year'] == 2018]
df_2019 = df[df['year'] == 2019]
```

```python
aumento = pd.Series((df_2019['value'].to_numpy() - df_2018['value'].to_numpy())*100/df_2018['value'].to_numpy())
```

$\Uparrow \Uparrow \Uparrow \Uparrow$ Na célula acima, por alguma razão, tive que usar o numpy para substrair as colunas termo a termo, pois o resultado vinha sendo NaN só com Series do pandas... $\Uparrow \Uparrow \Uparrow \Uparrow$

```python
novo_df = [[pais, valor] for pais, valor in zip(df_2018['country_name'].unique(), aumento)]
novo_df = pd.DataFrame(novo_df)
novo_df.columns = ['country_name', 'CO2 value']
```

```python
novo_df = novo_df.sort_values('CO2 value', ascending = False)
df_top10 = novo_df.iloc[0:10]
```

```python
df_top10.plot.barh(x = 'country_name', y = 'CO2 value')
```
